```python
print("Anusri T-24BAD010")
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv(r"/kaggle/input/ecommerce-data/data.csv",encoding="latin1")
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())
df['Sales'] = df['Quantity'] * df['UnitPrice']
top_products = df.groupby('Description')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure()
top_products.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
daily_sales = df.groupby(df['InvoiceDate'].dt.date)['Sales'].sum()
plt.figure()
plt.plot(daily_sales)
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()


```

```python
print("Anusri T-24BAD010")
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"/kaggle/input/housing-price-prediction/Housing.csv",encoding="latin1")
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())
df_clean = df.dropna()
plt.figure()
plt.scatter(df['area'], df['price'])
plt.title("Area vs Price")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.show()
plt.figure()
plt.scatter(df['bedrooms'], df['price'])
plt.title("Bedrooms vs Price")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()
plt.figure()
plt.scatter(df['bathrooms'], df['price'])
plt.title("Bathrooms vs Price")
plt.xlabel("Bathrooms")
plt.ylabel("Price")
plt.show()
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Housing Features")
plt.show()
df.groupby('bedrooms')['price'].mean()
df.groupby('airconditioning')['price'].mean()




```

```python
print("Anusri T-24BAD010")
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"/kaggle/input/customer-personality-analysis/marketing_campaign.csv",sep="\t")
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())
df['Age'] = 2024 - df['Year_Birth']
avg_spending = df.groupby('Education')['MntWines'].mean()
plt.figure()
avg_spending.plot(kind='bar')
plt.title("Average Wine Spending by Education")
plt.xlabel("Education")
plt.ylabel("Average Spending")
plt.show()
plt.figure()
sns.boxplot(x=df['Income'])
plt.title("Income Distribution")
plt.xlabel("Income")
plt.show()
spending_marital = df.groupby('Marital_Status')['MntWines'].mean()
plt.figure()
spending_marital.plot(kind='bar')
plt.title("Average Spending by Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Average Spending")
plt.show()
df.groupby('Response')['Income'].mean()
df.groupby('Response')['Age'].mean()



```
